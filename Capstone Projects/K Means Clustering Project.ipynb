### K Means Clustering 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
#fit model to all the data except Private label
kmeans.fit(college.drop('Private',axis=1))
```

```python
#display cluster vectors
kmeans.cluster_centers_
```

### Evaluate Model Since the Data Set Provides the Labels

#### Create a new column for df called 'Cluster', in which is a 1 for a Private school and a 0 for a public school.

```python
def converter(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
college['Cluster'] = college['Private'].apply(converter)
```

```python
college.head()
```

#### Create confusion matrix and classification report to see how the model worked without being given labels

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(college['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(college['Cluster'],kmeans.labels_))
```
